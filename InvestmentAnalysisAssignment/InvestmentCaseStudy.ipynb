```python
#Drop columns
loansDF = loansDF.drop(['tot_hi_cred_lim', 'total_bal_ex_mort', 'total_bc_limit', 'total_il_high_credit_limit', 'num_tl_op_past_12m','pct_tl_nvr_dlq', 'percent_bc_gt_75' ], axis=1)
loansDF = loansDF.drop(['num_tl_120dpd_2m', 'num_tl_30dpd', 'num_tl_90g_dpd_24m', 'num_actv_bc_tl', 'num_actv_rev_tl', 'num_bc_sats'], axis=1)
loansDF = loansDF.drop(['num_rev_accts', 'num_rev_tl_bal_gt_0', 'num_sats', 'num_bc_tl', 'num_il_tl', 'num_op_rev_tl'], axis=1)
loansDF = loansDF.drop(['mths_since_recent_inq', 'mths_since_recent_revol_delinq', 'num_accts_ever_120_pd'], axis=1)
loansDF = loansDF.drop(['mort_acc', 'mths_since_recent_bc', 'mths_since_recent_bc_dlq', 'mo_sin_old_rev_tl_op', 'mo_sin_rcnt_rev_tl_op', 'mo_sin_rcnt_tl'], axis=1)
loansDF = loansDF.drop(['mo_sin_old_il_acct', 'bc_util', 'bc_open_to_buy', 'avg_cur_bal', 'acc_open_past_24mths', 'inq_last_12m','total_cu_tl', 'inq_fi'], axis=1)
loansDF = loansDF.drop(['total_rev_hi_lim', 'all_util', 'max_bal_bc', 'open_rv_24m', 'open_rv_12m','il_util','total_bal_il','mths_since_rcnt_il' ], axis = 1)
loansDF = loansDF.drop(['open_il_24m', 'open_il_12m', 'open_il_6m', 'open_acc_6m', 'tot_cur_bal', 'tot_coll_amt'], axis = 1)
loansDF = loansDF.drop(['mths_since_last_major_derog', 'annual_inc_joint','dti_joint','verification_status_joint' ], axis = 1)
loansDF = loansDF.drop(['chargeoff_within_12_mths', 'tax_liens', 'delinq_amnt', 'acc_now_delinq', 'collections_12_mths_ex_med'], axis=1)
loansDF = loansDF.drop(['member_id'], axis=1)
loansDF = loansDF.drop(['next_pymnt_d', 'collection_recovery_fee', 'recoveries', 'total_pymnt', 'total_pymnt_inv', 'total_rec_prncp', 'total_rec_int', 'total_rec_late_fee'],axis=1)
loansDF = loansDF.drop(['revol_bal', 'revol_util', 'last_pymnt_d', 'last_pymnt_amnt', 'out_prncp', 'out_prncp_inv', 'mths_since_last_delinq', 'mths_since_last_record', 'policy_code', 'pub_rec_bankruptcies', 'initial_list_status', 'pub_rec'],axis=1)
loansDF = loansDF.drop(['pymnt_plan'],axis=1)
pd.set_option('display.max_rows', 10)
loansDF.isna().sum()
```

<b>Type Driven Metric- </b>

```python
# Type driven metric to change the type string into type float to analyse better 
loansDF['int_rate_percent'] = loansDF['int_rate'].apply(lambda x: float(x[:-1]))
loansDF['num_term_months'] = loansDF['term'].apply(lambda x:int(x.split(' months')[0]))
```

```python
# defining a few functions

def income_slab(x):
    if x < 50000:
        return '0-0.50L'
    elif x < 100000:
        return '0.50-1.00L'
    elif x < 150000:
        return '1.00-1.50L'
    elif x < 200000:
        return '1.50-2.00L'
    elif x < 250000:
        return '2.00-2.50L'       
    else:
        return '>=2.50L'

loansDF['annual_inc_slab'] = loansDF['annual_inc'].apply(income_slab)

def encodeStatus(loan_status):
    if loan_status == 'Charged Off':
        return 1
    #elif loan_status == 'Current':
    #    return 1
    else:
        return 0
loansDF['loan_status_code'] = loansDF['loan_status'].apply(encodeStatus)

def encode_dti(dti):
    if dti < 2.5:
        return '< 2.5'
    elif dti < 5:
        return '2.5 - 5.0'
    elif dti < 7.5:        
        return '5.0 - 7.5'    
    elif dti < 10:
        return '7.5 - 10.0'    
    elif dti < 12.5:
        return '10.0 - 12.5'
    elif dti < 15:
        return '12.5 - 15.0'
    elif dti < 17.5:
        return '15.0 - 17.5'    
    elif dti < 20:
        return '17.5 - 20.0'    
    else:
        return '> 20.0'
    
loansDF['dti_code'] = loansDF['dti'].apply(encode_dti)
```
